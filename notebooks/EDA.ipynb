# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
#Setting up SQL database

import sqlite3
from sqlite3 import Error

def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection

connection = create_connection("/Users/johnpayne/lighthouse-data-notes/Week_2/Day_4/mini-project-2/data/POI_Nashville.sqlite")

def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed succesfully")
    except Error as e:
        print(f"Query '{e}' not succesful")
    
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
```

Get the top 10 restaurants according to their rating

```python
#Creating merged table of both foursquare and yelp
create_response_table = """
CREATE TABLE IF NOT EXISTS response (
    response_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    category TEXT,
    review_count INTEGER,
    rating INTEGER,
    location TEXT,
    api TEXT
);    
"""
```

```python
execute_query(connection, create_response_table)
```
